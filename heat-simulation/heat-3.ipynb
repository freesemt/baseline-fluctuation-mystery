$$ \frac{\partial^2 u}{\partial x^2} - \frac{\partial u}{\partial t} = 0 $$
* $ T(x,t+k) = \lambda T(x-h, t) + (1 - 2\lambda) T(x,t) + \lambda T(x+h,t) $
* $ \lambda = \frac{k}{h^2} $
* $ T_{i,j} = \lambda T_{i-1,j-1} + (1 - 2\lambda) T_{i,j-1} + \lambda T_{i+1,j-1} $

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn
seaborn.set_theme()
%matplotlib widget
```

```python
h = 0.2
k = 0.01
x = np.arange(0, 100+h, h)
t = np.arange(0, 60+k, k)
n = len(x)
m = len(t)
boundaryConditions = [0, 0]
sigma = 0.1
mun = n//2
mu = x[mun]
initialConditions = 0.1*np.exp(-((x - mu)/sigma)**2)
# initialConditions = np.zeros(len(x))
# k = n//2
# initialConditions[k] = 0.1
# initialConditions[k-1:k+2] = 0.1
plt.plot(x, initialConditions)
print((n, m), mun, mu)
```

```python
T = np.zeros((n,m))
T[0,:] = boundaryConditions[0]
T[-1,:] = boundaryConditions[1]
T[:,0] = initialConditions
# T.round(3)
factor = k/h**2
for j in range(1,m):
    for i in range(1,n-1):
        T[i,j] = factor*T[i-1,j-1] + (1 - 2*factor)*T[i,j-1] + factor*T[i+1,j-1]

T.round(3)
T.shape
```

```python
T = np.zeros((n,m))
T[0,:] = boundaryConditions[0]
T[-1,:] = boundaryConditions[1]
T[:,0] = initialConditions
# T.round(3)
factor = k/h**2
for j in range(1,m):
    for i in range(1,n-1):
        T[i,j] = factor*T[i-1,j-1] + (1 - 2*factor)*T[i,j-1] + factor*T[i+1,j-1]

T.round(3)
T.shape
```

```python
R = np.linspace(1,0,m)
G = 0
B = np.linspace(0,1,m)
for j in range(0, m, 10):
    plt.plot(x, T[:,j], color=[R[j],G,B[j]], alpha=0.5)
# plt.legend(t.round(3))
plt.xlabel('distance [cm]')
plt.ylabel('Temperature [degree C]');
```

```python
u = mu/30
csigma = 10
def c(x, t):
    return np.exp(-((x - u*t)/csigma)**2)
```

```python
def plot_concentration_model(v):
    global u
    u = v
    fig, axes = plt.subplots(ncols=2, figsize=(12,6), subplot_kw=dict(projection='3d'))
    t_ = t
    lenx = len(x)
    lent = len(t_)
    for ax, view_dict, view_type in zip(axes, (dict(elev=15, azim=-15), dict(elev=15, azim=-75)), ("Positional", "Temporal")):
        ax.set_title("%s View" % view_type)
        for i in range(0, lenx, 50):
            xx = np.ones(lent)*x[i]
            zz = c(x[i], t_)
            alpha = 1 if i == 250 else 0.3
            label = 'x = %gcm' % x[i]  if i == 250 else None
            ax.plot(t_, xx, zz, color='C0', alpha=alpha, label=label, lw=2)
        for j in range(0, lent, 600):
            tt = np.ones(lenx)*t_[j]
            zz = c(x, t_[j])
            if j <= 3000:
                ls = '-'
                alpha = (j/3000)**2
            else:
                ls = ':'
                alpha = 0.8
            label = 't = %gsec' % t_[j] if j == 3000 else None
            ax.plot(tt, x, zz, ls=ls, color='C1', alpha=alpha, label=label, lw=2)
        ax.legend()
        ax.set_xlabel("time")
        ax.set_ylabel("position")
        ax.set_zlabel("concentration")
        ax.view_init(**view_dict)
    
    fig.tight_layout()
    fig.subplots_adjust(wspace=0.2)

plot_concentration_model(mu/30*1)
```
