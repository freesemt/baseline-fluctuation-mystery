* from [Solving Heat equation PDE using Explicit method in Python](https://www.youtube.com/watch?v=NLuCx2SrxHw)

$$ \frac{\partial^2 u}{\partial x^2} - \frac{\partial u}{\partial t} = 0 $$
* $ T(x,t+k) = \lambda T(x-h, t) + (1 - 2\lambda) T(x,t) + \lambda T(x+h,t) $
* $ \lambda = \frac{k}{h^2} $
* $ T_{i,j} = \lambda T_{i-1,j-1} + (1 - 2\lambda) T_{i,j-1} + \lambda T_{i+1,j-1} $

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
h = 0.025
k = 0.025
x = np.arange(0, 1+h, h).round(3)
t = np.arange(0, 0.1+k, k).round(3)
boundaryConditions = [0, 0]
initialConditions = np.sin(np.pi*x)

n = len(x)
m = len(t)
T = np.zeros((n,m))
T[0,:] = boundaryConditions[0]
T[-1,:] = boundaryConditions[1]
T[:,0] = initialConditions
T.round(3)
factor = k/h**2
for j in range(1,m):
    for i in range(1,n-1):
        T[i,j] = factor*T[i-1,j-1] + (1 - 2*factor)*T[i,j-1] + factor*T[i+1,j-1]

T.round(3)
n,m
```

```python
R = np.linspace(1,0,m)
G = 0
B = np.linspace(0,1,m)
for j in range(m):
    plt.plot(x, T[:,j], color=[R[j],G,B[j]])
plt.legend(t)
plt.xlabel('distance [m]')
plt.ylabel('Temperature [degree C]');
```
